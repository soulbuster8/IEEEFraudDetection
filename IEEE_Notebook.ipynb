```python
M_SERIES = ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'addr1', 'addr2', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6' ]
```

```python
def fill_missing_value_categorical(col, df):
    sorted_column = df[col].value_counts()
    df[col].fillna(sorted_column.index[0], inplace=True)
```

```python
[fill_missing_value_categorical(value, train_transaction) for value in M_SERIES]
```

```python
V_SERIES = ['V' + str(i) for i in range(1, 340)]
Other = ['dist1', 'dist2']
D_SERIES = ['D' + str(i) for i in range(1, 16)]
C_SERIES = ['C' + str(i) for i in range(1, 15)]
```

```python
features = []
features.extend(V_SERIES)
features.extend(Other)
features.extend(D_SERIES)
features.extend(C_SERIES)
```

```python
len(features)
```

```python
def fill_missing_value_numeric(col, df):
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)
```

```python
[fill_missing_value_numeric(value, train_transaction) for value in features]
```

```python
train_transaction.notnull().all().all()
```
